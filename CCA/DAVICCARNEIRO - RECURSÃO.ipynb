```python
def calcular_soma_recursiva(vetor, indice=0):
    if indice == len(vetor) - 1:
        return vetor[indice]
    else:
        return vetor[indice] + calcular_soma_recursiva(vetor, indice + 1)

# Exemplo de uso
vetor = [3, 7, 1, 9, 5, 2, 8]
soma_total = calcular_soma_recursiva(vetor)
print("Vetor:", vetor)
print("Soma total:", soma_total)

''''
Melhor Caso:

Tempo: O(n) - O melhor caso ocorre quando a função precisa percorrer todo o vetor, o que é inevitável para calcular a soma de todos os elementos. Portanto, a complexidade de tempo é linear, onde n é o número de elementos no vetor.
Espaço: O(n) - O espaço de pilha utilizado pela recursão depende do número de chamadas recursivas necessárias, o que é igual ao tamanho do vetor. Portanto, a complexidade de espaço é linear no pior caso.
Pior Caso:

Tempo: O(n) - O pior caso é o mesmo que o melhor caso, pois a função sempre precisa percorrer todo o vetor para calcular a soma.
Espaço: O(n) - O espaço de pilha utilizado pela recursão é linear e depende do tamanho do vetor.
Caso Médio:

Tempo: O(n) - O caso médio é dominado pelo mesmo fator que o melhor e pior caso, que é percorrer todo o vetor para calcular a soma.
Espaço: O(n) - O espaço de pilha utilizado pela recursão é linear e depende do tamanho do vetor.
'''
```
